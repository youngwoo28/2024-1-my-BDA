```python
!ls -al data/speech_moon.txt
```

```python
moon = open('data/speech_moon.txt', encoding = 'UTF-8').read()
moon
```

#### 2. 불필요한 문자 제거하기

```python
# 불필요한 문자 제거하기
import re #regular expression
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

###
-

```python
from konlpy.tag import Hannanum
hannanum = Hannanum()
print(hannanum.nouns(u'다람쥐 헌 쳇바퀴에 타고파'))
```

#### 3. 명사 추출하기

```python
# hannanum 만들기
import konlpy
hannanum = konlpy.tag.Hannanum()
```

```python
# 명사 추출하기, 기말시험
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
# 연설문에서 명사 추출하기
nouns = hannanum.nouns(moon)
# nouns
```

```python
type(nouns)
```

```python
len(nouns)
```

```python
nouns[:10]
```

```python
# 데이터 프레임으로 변환, 기말시험
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

#### 4. 단어 빈도표 만들기

```python
# 글자 수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
df_word[df_word['count'] >= 2]
```

```python
# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
## 단어 빈도 구하기

# 단어별 분리
# 빈도 구하기
# 내림차순 정렬
# 기말고사 예상문제

df_word = df_word.groupby('word', as_index = False) \
                 .agg(n = ('word', 'count')) \
                 .sort_values('n', ascending = False)
df_word
```

#### 5. 단어 빈도 막대 그래프 만들기

```python
# 단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

### 가장 많이 사용된 단어 알아보기

#### 1. 기사 댓글 불러오기

```python
# 데이터 불러오기
df = pd.read_csv('data/news_comment_BTS.csv', encoding = 'UTF-8')
```

```python
# 데이터 살펴보기
df.info()
```

```python
df.head()
```

```python
df['reply'][:10]
```

#### 2. 불필요한 문자 제거하기

```python
# 불필요한 문자 제거하기
df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True)
df['reply'].head()
```

#### 3. 명사 추출하기

```python
# kkma(꼬꼬마 형태소 분석기) 만들기
import konlpy
kkma = konlpy.tag.Kkma()
```

```python
from konlpy.tag import Kkma
from konlpy.utils import pprint
kkma = Kkma()
pprint(kkma.sentences(u'네, 안녕하세요. 반갑습니다.'))
pprint(kkma.nouns(u'질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요.'))
```

```python
# 명사 추출 - apply() 활용 : 오랜 시간 걸림
nouns = df['reply'].apply(kkma.nouns)
nouns
```

```python
type(nouns)
```

#### 4. 단어 빈도표 만들기

```python
# 한 행에 한 단어가 들어가도록 구성
nouns = nouns.explode()
nouns
```

```python
df = pd.DataFrame({'A': [[0, 1, 2], 'foo', [], [3, 4]],
                   'B': 1,
                   'C': [['a', 'b', 'c'], np.nan, [], ['d', 'e']]})
df
```

```python
df.explode('A') #기말고사
```

```python
# 데이터 프레임 만들기
df_word = pd.DataFrame({'word' : nouns})
df_word.head()
```

```python
# 글자 수 추가
df_word['count'] = df_word['word'].str.len()

# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word
```

```python
## 빈도표 만들기

# 단어별 분리
# 빈도 구하기
# 내림차순 정렬
df_word = df_word.groupby('word', as_index = False) \
                 .agg(n = ('word', 'count')) \
                 .sort_values('n', ascending = False)
df_word
```

#### 5. 단어 빈도 막대 그래프 만들기
